# フーリエ変換

```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt

n_sozai = 8
n_sample = 100
n_channels = 3
dt_data = 0.1 * 1e-3
clr = ["r","g","b"]

# 平均スペクトル格納用（素材 x センサー x 周波数）
avg_fft = []

for i in range(n_sozai):
    fft_accum = None
    for j in range(n_sample):
        # データ読み込み
        df = pd.read_table(glob.glob(path + "tactile_data/" + dir_name[i] + f"/data_{int(j+1)}_*")[0], header=None)
        df_np = df.to_numpy().T
        in_data_0 = df_np[:n_channels, 3000:8000]
        nt = in_data_0.shape[1]

        # FFT計算
        fft_val = np.fft.rfft(in_data_0, axis=1)      # shape: (3, nt/2+1)
        fft_mag = np.abs(fft_val) / nt

        if fft_accum is None:
            fft_accum = fft_mag
        else:
            fft_accum += fft_mag
    
    # サンプルで割って平均
    avg_fft.append(fft_accum / n_sample)

avg_fft = np.array(avg_fft)  # shape: (素材数, センサー数, 周波数)

# 周波数軸
f = np.fft.rfftfreq(nt, dt_data)

# 素材ごとの平均スペクトルを表示（3センサー分）
fig, axes = plt.subplots(2, 4, figsize=(20,10), squeeze=False)
for i in range(n_sozai):
    for k in range(n_channels):
        axes[i//4, i%4].plot(f, avg_fft[i,k,:], color=clr[k], label=f"sensor {k+1}")
    axes[i//4, i%4].set_xlim([0, 600])
    axes[i//4, i%4].set_ylim([0, 0.001])
    axes[i//4, i%4].set_title(dir_name[i])
    axes[i//4, i%4].set_xlabel("Frequency [Hz]")
    axes[i//4, i%4].set_ylabel("Amplitude")
    axes[i//4, i%4].legend(fontsize=6)
plt.tight_layout()
plt.show()

```

```python
def calc_meissner(data, t, dt):

	I = np.zeros((4, len(t)))
	
	for i in range(len(t)):
		if(i!=0):
			dF_dt = np.abs(data[i] - data[i-1])/(t[i]-t[i-1])

			I[0, i] = I[0, i-1] + 1*dF_dt + (-I[0, i-1]*dt/(8*1* 1e-3)) 
			I[1, i] = I[1, i-1] + 0.24*dF_dt + (-(I[1, i-1] - 0.24*0.13)*dt/(200* 1e-3) )
			I[2, i] = I[2, i-1] + 0.07*dF_dt + (-I[2, i-1]*dt/(1744.6* 1e-3))
			I[3, i] = I[0, i] 

	return I[3,:]


def calc_merkel(data, t, dt):

	I = np.zeros((4, len(t)))
	
	for i in range(len(t)):
		if(i!=0):
			dF_dt = np.abs(data[i] - data[i-1])/(t[i]-t[i-1])

			if(dF_dt < 0):
				dF_dt = 0
			I[0, i] = I[0, i-1] + 0.74*dF_dt + (-I[0, i-1]*dt/(8*1* 1e-3)) 
			I[1, i] = I[1, i-1] + 0.24*dF_dt + (-(I[1, i-1] - 0.24*0.13)*dt/(200*1* 1e-3) )
			I[2, i] = I[2, i-1] + 0.07*dF_dt + (-I[2, i-1]*dt/(1744.6*1* 1e-3))
			I[3, i] = I[0, i] +  I[1, i] +  I[2, i] 

	return I[3,:]
```

```python
from brian2 import *

dt_data = 0.1e-3  # 0.1 ms
dt_ms   = dt_data * 1e3

# Meissner 用
eqs_meissner = '''
dI0_meiss/dt = (1.0/dt_data) * dFdt_ta(t, i) - I0_meiss/(8*ms) : 1
I_meissner = I0_meiss : 1
dt_data : second (constant)
'''

# Merkel 用
eqs_merkel = '''
dJ0_mer/dt   = (0.74/dt_data) * dFdt_ta(t, i) - J0_mer/(8*ms)                 : 1
dJ1_mer/dt   = (0.24/dt_data) * dFdt_ta(t, i) - (J1_mer - 0.24*0.13)/(200*ms) : 1
dJ2_mer/dt   = (0.07/dt_data) * dFdt_ta(t, i) - J2_mer/(1744.6*ms)            : 1
I_merkel   = J0_mer + J1_mer + J2_mer : 1
dt_data : second (constant)
'''

fig1,axes1 = plt.subplots(2,4,figsize=(20,10),squeeze=False)
fig2,axes2 = plt.subplots(2,4,figsize=(20,10),squeeze=False)
fig3,axes3 = plt.subplots(2,4,figsize=(20,10),squeeze=False)

N_sensors = 3

for i in range(8):
    for j in range(1):    # サンプル 1 個だけ
        df = pd.read_table(
            glob.glob(path+"tactile_data/" +dir_name[i]+ f"/data_{int(j+1)}_*")[0],
            header=None
        )
        df_np = df.to_numpy().T
        in_data_0 = df_np[:3, 3000:8000]   # (3, nt)
        nt = in_data_0.shape[1]
        t_array = np.arange(nt) * dt_data

        # ---- 元の calc_* と同じ dF_dt (= |ΔF| / dt_data) を作る ----
        dF_dt_all = np.zeros((nt, N_sensors))
        for k in range(N_sensors):
            data_k = in_data_0[k, :]
            dF_dt_all[1:, k] = np.abs(data_k[1:] - data_k[:-1]) / dt_data
            dF_dt_all[0, k]  = dF_dt_all[1, k]

        # ---- Brian2 ----
        start_scope()
        defaultclock.dt = dt_ms*ms

        # 時間×センサの TimedArray（単位なし）
        dFdt_ta = TimedArray(dF_dt_all, dt=dt_ms*ms)

        # Meissner フィルタ用グループ
        G_meissner = NeuronGroup(N_sensors, eqs_meissner, method='euler')
        G_meissner.dt_data = dt_data * second
        G_meissner.I0_meiss = 0

        # Merkel フィルタ用グループ
        G_merkel = NeuronGroup(N_sensors, eqs_merkel, method='euler')
        G_merkel.dt_data = dt_data * second
        G_merkel.J0_mer = 0
        G_merkel.J1_mer = 0
        G_merkel.J2_mer = 0

        # モニタ
        M_meiss = StateMonitor(G_meissner, 'I_meissner', record=True)
        M_mer   = StateMonitor(G_merkel,   'I_merkel',   record=True)

        # シミュレーション実行
        run(nt*dt_ms*ms)

        # ---- input_current を作る ----
        input_current = np.zeros((6, nt))
        for k in range(N_sensors):
            meissner_k = np.array(M_meiss.I_meissner[k])
            merkel_k   = np.array(M_mer.I_merkel[k])

            input_current[2*k,   :] = 0.01 * merkel_k
            input_current[2*k+1, :] = 0.04 * meissner_k

        # ここでは ch0, ch1, ch2 の Merkel を描いてみる例
        axes1[i//4, i%4].plot(t_array, input_current[0])  # ch0 Merkel
        axes2[i//4, i%4].plot(t_array, input_current[2])  # ch1 Merkel
        axes3[i//4, i%4].plot(t_array, input_current[4])  # ch2 Merkel

    axes1[i//4, i%4].set_title(dir_name[i], fontsize=22)
    axes2[i//4, i%4].set_title(dir_name[i], fontsize=22)
    axes3[i//4, i%4].set_title(dir_name[i], fontsize=22)

fig1.text(0.5, 0.04, "Time [s]", ha="center", fontsize=22)
fig1.text(0.07, 0.5, "Voltage [V]", va="center", rotation="vertical", fontsize=22)
fig2.text(0.5, 0.04, "Time [s]", ha="center", fontsize=22)
fig2.text(0.07, 0.5, "Voltage [V]", va="center", rotation="vertical", fontsize=22)
fig3.text(0.5, 0.04, "Time [s]", ha="center", fontsize=22)
fig3.text(0.07, 0.5, "Voltage [V]", va="center", rotation="vertical", fontsize=22)

plt.show()
```
